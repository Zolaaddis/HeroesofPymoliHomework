## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
males_df = df.loc[df['Gender'] == "Male",:]
females_df = df.loc[df['Gender'] == "Female",:]

other_list = []
for x in range(0, len(df['Gender'])):
    if(df['Gender'][x] != "Male" and df['Gender'][x] != "Female"):
       other_list.append(df['Gender'][x])

other_set = set(other_list)
other_df = df.loc[df['Gender'] == other_list[0],:]
    
female_players = len(females_df['SN'].unique())
male_players = len(males_df['SN'].unique())
other_players = len(other_df['SN'].unique())
player_gender_counts_list = []
gender_percentages_list = []

player_gender_counts_list.append(female_players)
player_gender_counts_list.append(male_players)
player_gender_counts_list.append(other_players)

for y in range(0, len(player_gender_counts_list)):
    gender_percentages_list.append(player_gender_counts_list[y]/unique_players * 100)

gender_demographics_dict = {
    "Gender": ["Female", "Male", "Other / Non-Disclosed"],
    "Number of Players": player_gender_counts_list, "Percentage of Players": gender_percentages_list}

gender_demographics_df = pd.DataFrame(gender_demographics_dict)
gender_demographics_df["Percentage of Players"] = gender_demographics_df["Percentage of Players"].map("{:,.2f}".format)
gender_demographics_df
```

* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points

```python
# Establish bins for ages
bins = [0, 9, 14, 19, 24,29, 34, 39, 1000]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

# Categorize the values of the"Age" column into bins
purchase_data["Age group"] = pd.cut(purchase_data["Age"],
                                   bins,labels=group_names)

age_grouped = purchase_data.groupby("Age group")
unique_members = age_grouped["SN"].nunique()

age_demo = pd.DataFrame({"Total Counts":age_grouped["SN"].nunique(),"Percentage of Players":100*unique_members/unique_players})

age_demo["Percentage of Players"] = age_demo["Percentage of Players"].map("{:.2f}".format)

age_demo
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins = [0,9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99.99]
group_names = ["<10","10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Create a copy of purchase_data DataFrame
age_df = df.loc[:,:]

# Categorize the values of the"Age" column into bins
age_df ["Age Range"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

age_group = age_df.groupby('Age Range')

# Find the number of purchases per age range
age_purchase_count = age_group["Purchase ID"].count()

# Find the average purchase price per age range
avg_price_per_age = age_group["Price"].mean()

# Find the total purchase value per age range
total_value_per_age = age_group["Price"].sum()

# Find the average purchase total per person in an age range
avg_total_by_age = age_group["Price"].sum()/age_purchase_count

# Create a DataFrame 
age_ranges_analysis = pd.DataFrame({"Total Count": age_purchase_count,
                                    "Average Purchase Price": avg_price_per_age,
                                    "Total Purchase Value": total_value_per_age,
                                    "Average Purchase Total per Person": avg_total_by_age})

age_ranges_analysis["Average Purchase Price"] = age_ranges_analysis["Average Purchase Price"].map("${:,.2f}".format)
age_ranges_analysis["Total Purchase Value"] = age_ranges_analysis["Total Purchase Value"].map("${:,.2f}".format)
age_ranges_analysis["Average Purchase Total per Person"] = age_ranges_analysis["Average Purchase Total per Person"].map("${:,.2f}".format)

age_ranges_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group the original DataFrame by screenname
top_spenders = purchase_data.groupby("SN")

# Create a Series with "SN" as the index and purchase count as the value
spender_purchase_count = top_spenders["Price"].count()

# Find the total amount spent for each screenname
spender_total = top_spenders["Price"].sum()

# Calculate the average price spent 
spender_avg_price = spender_total/spender_purchase_count

# Create a DataFrame with calculated values
spender_df = pd.DataFrame({"Purchase Count": spender_purchase_count,
                          "Average Purchase Price": spender_avg_price,
                          "Total Purchase Value": spender_total})

# Sort DataFrame by "Total Purchase Value" column in descending order
spender_df.sort_values(by=['Total Purchase Value'], ascending=False, inplace=True)

spender_df["Average Purchase Price"] = spender_df["Average Purchase Price"].map("${:,.2f}".format)
spender_df["Total Purchase Value"] = spender_df["Total Purchase Value"].map("${:,.2f}".format)

spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a DataFrame 
item_df = purchase_data.loc[:, ['Item ID', 'Item Name', 'Price']]

# Create a groupby object 
item_group = item_df.groupby(["Item ID", "Item Name"])

# Find the number of times the item was purchased
item_count = item_group['Price'].count()

# Find the total amount spent on the item
item_total = item_group['Price'].sum()

item_price = item_total/item_count

# Create a DataFrame displaying the above values
item_summary = pd.DataFrame({"Purchase Count": item_count,
                             "Item Price": item_price,
                             "Total Purchase Value": item_total})

# Sort the DataFrame by "Purchase Count" column in descending order 
item_sorted = item_summary.sort_values(by=['Purchase Count'], ascending=False)

# Formating
item_sorted["Item Price"] = item_sorted["Item Price"].map("${:,.2f}".format)
item_sorted["Total Purchase Value"] = item_sorted["Total Purchase Value"].map("${:,.2f}".format)

item_sorted.head()
```
